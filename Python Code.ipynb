```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_excel("Task 1.xlsx")

months = data["Month"]
city_a_temp = data["City A Temp (Celsius)"]
city_b_temp = data["City B Temp (Celsius)"]


plt.figure(figsize=(10, 6))


plt.plot(months, city_a_temp, marker='o', label='City A', color='blue')
plt.plot(months, city_b_temp, marker='s', label='City B', color='red')


plt.xlabel("Months")
plt.ylabel("Temperature (C)")
plt.title("Monthly Temperature Comparison between City A and City B (2024)")

plt.legend()

plt.grid(True)
plt.show()
```

**Task 1 Report**
May and June has the highest temperatures making them ideal months for hosting matches. This ensure comfortable conditions.

```python
data3 = pd.read_excel("Task 3.xlsx")

hours_of_study = data3["Hours of Study"]
exam_scores = data3["Exam Score"]

plt.figure(figsize=(8, 6))
plt.scatter(hours_of_study, exam_scores, color='blue', alpha=0.7, label='Data Points')

model = LinearRegression()
model.fit(hours_of_study.values.reshape(-1, 1), exam_scores.values.reshape(-1, 1))
plt.plot(hours_of_study, model.predict(hours_of_study.values.reshape(-1, 1)), color='red', label='Trend Line')

plt.xlabel('Hours of Study')
plt.ylabel('Exam Score')

plt.title('Relationship between Hours of Study and Exam Scores')

plt.legend()

plt.grid(True)
plt.show()
```

**Task Report 3** The scatter plot graph shows a positive correlation between study hours and exam scores. As study hours increase exam scores tend to rise.
Students who study for fewer hours (closer to the left side of the graph) generally score lower on exams.
Conversely, those who invest more time on study perform better.
While this graph doesnt prove causation, it highlights the importance of consistent study habits for attaining good grades in exams.
